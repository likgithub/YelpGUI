```python
"""

Get relevant stats for question 2

RAW:
- average star rating
- comments made by user
- yelping since
GENERATED STATS mean, median, mode, 25p, 75p:
- review count
- useful, funny, cool, fans
- compliments: hot, more, profile, cute, list, note, plain, cool, funny, writer, photos


"""
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv('user_all.csv', low_memory=False))
```

```python
for c in df.columns: print(c)
```

```python
#for col in df.columns: print(col)

compliments = [c for c in df.columns[11:len(df.columns)+1]]
general = ['review_count', 'useful', 'funny', 'cool', 'fans', 'average_stars']
```

```python
df_user_stats= df[general + compliments]
```

```python
for c in df_user_stats.columns: print(c)
```

```python
df_big_four = pd.DataFrame({'column': general + compliments,
                            'mean': df_user_stats.mean(),
                            'median': df_user_stats.median(),
                            'Percentile25': df_user_stats.quantile(0.25),
                            'Percentile75': df_user_stats.quantile(0.75)})

df_big_four.reset_index(drop=True, inplace=True)
df_big_four.to_csv(r'user/user_stats.csv', index=False, encoding='utf-8')
```
